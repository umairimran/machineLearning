```python
df=pd.read_csv("/kaggle/input/mushroom-classification/mushrooms.csv")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import CategoricalNB  # For categorical features
from sklearn.preprocessing import OrdinalEncoder
from sklearn.metrics import accuracy_score, classification_report
import pandas as pd
```

```python
x=df.drop(columns=["class"])
y=df["class"]
```

```python
encoder = OrdinalEncoder()
x = encoder.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
classifier=CategoricalNB()
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
accuracy=accuracy_score(y_test,y_pred)
```

```python
def predict_class(features):
    # Ensure features are in the correct format (2D array for sklearn)
    features = [features]
    # Encode features using the same encoder
    encoded_features = encoder.transform(features)
    # Predict the class using the trained model
    prediction = classifier.predict(encoded_features)
    return prediction[0]
```

```python
features_list = ['x', 's', 'n', 't', 'p', 'f', 'c', 'n', 'k', 'e', 'e', 's', 's', 'w', 'w', 'p', 'w', 'o', 'p', 'k', 's', 'u']
predict_class(features_list)
```
